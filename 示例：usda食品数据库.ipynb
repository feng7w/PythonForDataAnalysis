现在，为了对全部营养数据做一些分析，
最简单的办法是将所有食物的营养成分整合到一个大表中。

我们分几个步骤来实现该目的。

首先，将各食物的营养成分列表转换为一个DataFrame，并添加一个表示编号的列，

然后，将该DataFrame添加到一个列表中，

最后，通过concat将这些东西连接起来就可以了：

```python
nutrients = []
for rec in db:
    fnuts = pd.DataFrame(rec['nutrients'])
    fnuts['id'] = rec['id']
    nutrients.append(fnuts)
    
nutrients = pd.concat(nutrients, ignore_index=True)
```

nutrients应该是下面这样的：

```python
nutrients
```

删除重复值：

```python
#重复的数量
nutrients.duplicated().sum() 
```

```python
nutrients = nutrients.drop_duplicates()
```

由于两个DataFrame对象中都有"group"和"description"，所以需要对它们进行重命名：

```python
col_mapping = {'description' : 'food', 
               'group' : 'fgroup'}
```

```python
info = info.rename(columns=col_mapping, copy=False)
```

```python
info.info()
```

```python
col_mapping = {'description' : 'nutrient', 
               'group' : 'nutgroup'}
```

```python
nutrients = nutrients.rename(columns=col_mapping, copy=False)
```

将info跟nutrients合并起来：

```python
ndata = pd.merge(nutrients, info, on='id', how='outer')
```

```python
ndata.info()
```

```python
ndata.iloc[3000]
```

根据食物分类和营养类型画出一张中位值图：
（根据营养分类得出的锌中位值）

```python
result = ndata.groupby(['nutrient', 'fgroup'])['value'].quantile(0.5)
result['Zinc, Zn'].sort_values().plot(kind='barh')
```

发现各营养成分最为丰富的食物是什么：

```python
by_nutrient = ndata.groupby(['nutgroup', 'nutrient'])
```

```python
get_maximun = lambda x: x.loc[x.value.idxmax()]
get_minimun = lambda x: x.loc[x.value.idxmin()]
```

```python
max_foods = by_nutrient.apply(get_maximun)[['value', 'food']]
```

```python
# 让food小一点
max_foods.food = max_foods.food.str[:50]
```
